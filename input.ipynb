## Cách nhập code bài 1
### Dòng 1 nhập số đỉnh (bài này số đỉnh là 7)
### Dòng 2 nhập theo format: đỉnh_gốc  đỉnh_kề1  trọng_số1 ...  đỉnh_kề(n) trọng_số(n)
1 2 9 4 16 7 17 3 18  
2 1 9 6 10 4 30 3 27  
3 5 11 1 18 4 19 2 27  
4 1 16 2 30 3 19 5 28  
5 3 11 4 28 7 30  
6 2 10 7 25  
7 1 17 6 25 5 30  
### Nhấn Enter là ra kết quả
